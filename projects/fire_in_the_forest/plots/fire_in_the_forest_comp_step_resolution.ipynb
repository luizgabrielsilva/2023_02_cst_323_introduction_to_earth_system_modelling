# Experiment comparison

```python
from matplotlib import pyplot as plt

import pathlib
from experiment import Experiment
```

```python
project_path = pathlib.Path().parent.resolve().parent
```

```python
# Parameters

exp1_params = dict(
    experiment_name="Exp1 - FinalTime=150",
    filename="experiment-1-limited-step001-ft150.csv",
    project_path=project_path,
    dimension=50,
    color="red",
    expected_forest=True,
)

exp2_params = dict(
    experiment_name="Exp1 - Verification",
    filename="experiment-1-tests-04-06.csv",
    project_path=project_path,
    dimension=50,
    color="blue",
    expected_forest=False,
)


experiments = [exp1_params,
               #exp2_params,
               #exp3_params,
               #exp4_params,
               exp5_params]
```

```python

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5))

for experiment in experiments:
    Exp = Experiment(experiment_name=experiment["experiment_name"],
                     project_path=experiment["project_path"],
                     filename=experiment["filename"],
                     dimension=experiment["dimension"])
    
    Exp.get_data()
    Exp.process_data()
    
    Exp.plot_results(expected_forest=experiment["expected_forest"],
                     color=experiment["color"],
                     label=experiment["experiment_name"],
                     ci=True,
                     ax=ax)
    
    
```
