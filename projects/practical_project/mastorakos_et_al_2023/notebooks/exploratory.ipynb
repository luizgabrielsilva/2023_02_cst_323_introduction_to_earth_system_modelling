```python
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
sample_size = 2000
lower_threshold = 0
higher_threshold = 2 * np.pi

epsilon = np.random.uniform(low=lower_threshold, high=higher_threshold, size=sample_size)
X = np.cos(epsilon)
Y = np.sin(epsilon)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))


sns.histplot(x=X, y=Y, bins="auto", ax=ax)


plt.show()
```
