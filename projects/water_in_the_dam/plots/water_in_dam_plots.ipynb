# Water in the Dam - Plots


Notebook utilizado para plotar os gráficos das simulações

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

import pathlib
```

```python
project_path = pathlib.Path().parent.resolve().parent
data_path = project_path/ pathlib.Path("data")
```

## Precipitação

São Paulo - Estação Mirante de Santana 

Normal climatológica 1991-2020 -
https://clima.inmet.gov.br/GraficosClimatologicos/SP/83781


São José dos Campos

Normal Climatológica 1961-1990 - 
https://clima.inmet.gov.br/GraficosClimatologicos/SP/83809


```python
data_sao_paulo = pd.read_csv(data_path / pathlib.Path("pluviometry/sao_paulo_pluviometria.csv"))
data_sao_jose = pd.read_csv(data_path / pathlib.Path("pluviometry/sao_jose_dos_campos_pluviometria.csv"))
```

```python
data_sao_paulo.head()
```

```python
data_sao_jose.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 5))

ax.plot(data_sao_paulo["Month"],
        data_sao_paulo["%"],
        "--o",
        color="blue",
        markerfacecolor="none",
        label="São Paulo (1991-2020)")

ax.plot(data_sao_jose["Month"],
        data_sao_jose["%"],
        "--o",
        color="red",
        markerfacecolor="none",
        label="São José dos Campos (1961-1990)")

ax.set_xlabel("Mês")
ax.set_ylabel("Fração da precipitação")

ax.legend()

plt.show()
```

## Testes

```python
first_year = 15
last_year = 30

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,5))

for i, j, c in zip(["0.1", "0.2", "1.1"], [1, 2, 12], ["blue", "green", "red"]):
    tmp = pd.read_csv(data_path / pathlib.Path("water_in_dam_v" + i + "/waterDam_v_" + i + "_sim_1.csv"),
                      header=0, names=["water_vol"])

    ax.plot(tmp.index[first_year * j : last_year * j] / j + 1950,
            tmp.water_vol[first_year * j : last_year * j], "--o",
            markerfacecolor="none", label="Water Dam " + i, linewidth=0.8,
            markersize=3, color=c)


ax.legend()
ax.set_ylabel("Water Volume in the Dam (m³)")
ax.set_xlabel("Year")

plt.show()
```

## Simulações

```python
ref_year = 1950

colors = [
    "violet",
    "blue",
    "green",
    "orange",
    "red",
]

first_year = 15
last_year = 60

sim_time = [30, 35, 60, 25, 40]
```

### Water In the Dam v0.1

```python
m = 1
version = "0.1"

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,5))

for sim, time in zip(range(0, 5), sim_time):
    tmp = pd.read_csv(data_path / pathlib.Path("water_in_dam_v" + version
                                               + "/waterDam_v_" + version + "_sim_" + str(sim + 1) + ".csv"),
                      header=0, names=["water_vol"])

    ax.plot(tmp.index[first_year * m : time * m] / m + ref_year,
            tmp.water_vol[first_year * m : time * m], "--o",color=colors[sim],
            markerfacecolor="none", label="Sim " + str(sim + 1), linewidth=0.5, markersize=4)


ax.legend()
ax.set_ylabel("Water Volume in the Dam (m³)")
ax.set_xlabel("Year")

plt.show()
```
